## Task 1: Summarize

**Objective**: Generate concise summaries of text while preserving key information.


```python
# !pip install openai

import openai
import os
from dotenv import load_dotenv

load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

# Initialize the client
client = openai.OpenAI(api_key=OPENAI_API_KEY)
model_version = "gpt-3.5-turbo"

if not client:
    raise ValueError("OPENAI_API_KEY not found")

def get_llm_response(prompt, model=model_version, max_tokens=500):
    """
    Send a prompt to ChatGPT and get the response.
    """
    try:
        response = client.chat.completions.create(
            model=model,
            messages=[{"role": "user", "content": prompt}],
            max_tokens=max_tokens,
            temperature=0.7  # Adjust for creativity (0-1, lower = more focused)
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}\n\nMake sure you've set your OPENAI_API_KEY!"

print("✅ OpenAI API setup complete!")
```

```python
'''
A page from Chinua Achebe's Things Fall Apart
'''

input_text = """
Okonkwo was well known throughout the nine
villages and even beyond. His fame rested on
solid personal achievements. As a young man of eighteen he
had brought honor to his village by throwing Amalinze the
Cat. Amalinze was the great wrestler who for seven years was
unbeaten, from Umuofia to Mbaino. He was called the Cat
because his back would never touch the earth. It was this
man that Okonkwo threw in a fight which the old men
agreed was one of the fiercest since the founder of their town
engaged a spirit of the wild for seven days and seven nights.

The drums beat and the flutes sang and the spectators held
their breath. Amalinze was a wily craftsman, but Okonkwo
was as slippery as a fish in water. Every nerve and every
muscle stood out on their arms, on their backs and their
thighs, and one almost heard them stretching to breaking
point. In the end Okonkwo threw the Cat.

That was many years ago, twenty years or more, and
during this time Okonkwo's fame had grown like a bush-fire
in the harmattan. He was tall and huge, and his bushy
eyebrows and wide nose gave him a very severe look. He
breathed heavily, and it was said that, when he slept, his
wives and children in their houses could hear him breathe.
When he walked, his heels hardly touched the ground and he
seemed to walk on springs, as if he was going to pounce on
somebody. And he did pounce on people quite often. He had
a slight stammer and whenever he was angry and could not
get his words out quickly enough, he would use his fists. He
had no patience with unsuccessful men. He had had no
patience with his father.
"""

print(input_text)
```

### Variant 1: Direct Instruction

Simple, straightforward prompt asking for a summary.


```python
prompt_1 = f"Summarize the following text:\n\n{input_text}"

print("OUTPUT:")
print("\n" + "-"*50)

output_1 = get_llm_response(prompt_1)
print(output_1)
```

#### Evaluation

- **Fluency** (1-5): 5/5 - Clear and readable, no grammer issue
- **Faithfulness** (1-5): 5/5 - Accurately captures all key points from source
- **Usefulness** (1-5): 4/5 - Good summary, could be more concise


### Variant 2: Length Constraint

Asking for a specific length (2 sentences) to control output size.


```python
prompt_2 = f"Summarize the following text in exactly 2 sentences:\n\n{input_text}"

print("OUTPUT:")
print("\n" + "-"*50)

output_2 = get_llm_response(prompt_2)
print(output_2)
```

#### Evaluation

- **Fluency** (1-5): 5/5 - Well-structured, flows naturally
- **Faithfulness** (1-5): 5/5 - All main points preserved
- **Usefulness** (1-5): 4/5 - Concise and informative but feels lacking


### Variant 3: Bullet Points Format

Requesting output in a structured bullet point format.


```python
prompt_3 = f"Summarize the following text using bullet points:\n\n{input_text}"

print("OUTPUT:")
print("\n" + "-"*50)

output_3 = get_llm_response(prompt_3)
print(output_3)
```

#### Evaluation

- **Fluency** (1-5): 5/5 - Clear bullet structure
- **Faithfulness** (1-5): 5/5 - All key information included
- **Usefulness** (1-5): 5/5 - Easy to scan and understand


### Variant 4: Role-Based (Executive Summary)

Using a role (business analyst) to guide the output style.


```python
prompt_4 = f"You are a brilliant writer. Write an executive summary of the following text:\n\n{input_text}"


print("OUTPUT:")
print("\n" + "-"*50)

output_4 = get_llm_response(prompt_4)
print(output_4)
```

#### Evaluation

- **Fluency** (1-5): 5/5 - Well-written, flows smoothly, and reads naturally.
- **Faithfulness** (1-5): 3/5 - Includes core facts, omits some key info
- **Usefulness** (1-5): 4/5 - Readable narrative summary with good context


### Variant 5: Few-Shot Example

Providing an example to guide the output style.


```python
prompt_5 = f"""
Example:
Text: The weather today is sunny with temperatures reaching 75°F. Many people are enjoying outdoor activities.
Summary: Today's weather is sunny and warm (75°F), encouraging outdoor activities.

Now summarize this text:
{input_text}
"""

print("OUTPUT:")
print("\n" + "-"*50)

output_5 = get_llm_response(prompt_5)
print(output_5)
```

#### Evaluation

- **Fluency** (1-5):  5/5 – Clear, concise, and grammatically correct.
- **Faithfulness** (1-5): 4/5 - Accurate representation. Missing some specifics.
- **Usefulness** (1-5): 5/5 - Concise summary that covers the essential traits and is easy to scan.


### Variant 6: Structured Output (Key Points + Summary)

Requesting both key points and a summary paragraph.


```python
prompt_6 = f"""
Summarize the following text. Provide:
1. Key points (3-4 bullet points)
2. A brief summary paragraph

Text:
{input_text}
"""

print("OUTPUT:")
print("\n" + "-"*50)

output_6 = get_llm_response(prompt_6)
print(output_6)
```

#### Evaluation

- **Fluency** (1-5): 5/5 - Well-structured, clear format
- **Faithfulness** (1-5): 3/5 – Captures main points with missing specifics.
- **Usefulness** (1-5): 3/5 – Too brief; omits important details.
